```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lo primero que hacemos es cargar el dataframe que vamos a usar para esta lección
df = pd.read_csv("files/world_data_full_apply_nulos.csv", index_col=0) 


pd.set_option('display.max_columns', None)
```

```python
df.head()
```

1.- ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

```python
sns.histplot(data=df, x='birth_rate', bins=10, kde=False, color="purple",cumulative=True)

plt.xlabel("Tasa de natalidad")
plt.ylabel("Conteo")
plt.title("Distribución de natalidad", fontsize = 8)
```

2.- ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

```python
# creamos una gráfica básica usando 'sns.scatterplot()' 
# como ya lo hemos aprendido, vamos a cambiar el nombre de los ejes usando los métodos 'plt.xlabel()'  y 'plt.ylabel()' 
sns.scatterplot(x = "infant_mortality", 
                y = "life_expectancy", 
                data = df, color="purple",size=2)

# cambiamos los nombres de los ejes
plt.xlabel("Tasa mortalidad infantil")
plt.ylabel("Esperanza de vida")
plt.title("Relacion ejercicio 2 ", fontsize=8);
```

```python
"""Se observa una relación en los países que hay mayor esperanza de vida, existe menor tasa de mortalidad infantil"""
```

3.- ¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos.

```python
sns.boxplot(x = 'life_expectancy', data = df[df['continent']== "Europe"], width = 0.5, color = 'turquoise')

plt.xlabel("Esperanza vida Europa")

plt.title("Relacion ejercicio 3 ", fontsize=8)
```


4.-¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.

```python



# Crear la figura y los subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Relaciones entre la Tasa de Natalidad y otras variables')

# Lista de variables y títulos
variables =["life_expectancy",'density','infant_mortality', 'urban_population']

titulos = ['Tasa de Natalidad vs Tasa de Mortalidad Infantil','Tasa de Natalidad vs Esperanza de Vida','Tasa de Natalidad vs Densidad de Población','Tasa de Natalidad vs Población Urbana']

 # Bucle for para iterar

for ax, var, title, in zip(axes.flatten(), variables, titulos):

    #axes.flatten()--> sirve para recorrer la matrix 2,2
    #esto recorre 3 listas al mismo tiempo, la de variables, la de titulos
    #zip es para agrupar esyos elementos para obtener en cada iteracion: un ax, la variable del y y el título.

    sns.regplot(x=df['birth_rate'], y=df[var], ax=ax, scatter_kws={'color':'orange'}, line_kws={'color':'purple'})
    ax.set_title(title)
    ax.set_xlabel('Tasa de natalidad')
    ax.set_ylabel(var)

plt.show()
```
